# Text data and Image data

```python
demo = ['Number of covid19 cases today', 'How many covid19 cases today',
       'Number of covid19 patients cured', 'Covid19 Pandemic', 'number of cases in India', 
       'Cases in Bangalore']
```

```python
demo
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer()
x = cv.fit_transform(demo)
```

```python
x
```

```python
len(demo)
```

#Sparse matriz  - most of the values will be zero

```python
#onverting it to matrix form from compressed sparse Row format
x.todense()
```

```python
cv.get_feature_names()
```

```python
import pandas as pd
pd.DataFrame(x.todense(), columns = cv.get_feature_names())
```

# TFIDF matrix

```python
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(demo)
```

```python
x.todense()
```

```python
pd.DataFrame(x.todense(), columns = tfidf.get_feature_names())
```

```python
pd.DataFrame(x.todense(), columns = tfidf.get_feature_names()).shape
```

#Look into Cat folder and dog folder and import all images at one stance

```python
import glob
```

```python
#after path, ** this for sub folder and *.jog for each file
#image name along with the path
images = glob.glob('D:\Jigsaw\Classroom content\MLAI\Feature engg - Unstructured data\Cats and Dogs\**\*.jpg',
                  recursive = True)
images
```

```python
#import multiple images, resize and extract pixel information of images
pixels = []
for i in images:
    pix = cv2.resize(cv2.cvtColor(cv2.imread(i), cv2.COLOR_BGR2GRAY), (50,50), cv2.INTER_LINEAR).reshape(50*50)
    pixels.append(pix)
```

```python
pixels
```

```python
pd.DataFrame(pixels)
```

```python
#pixels value is 0 to 255
# normalizing it
data = pd.DataFrame(pixels)/255
```
